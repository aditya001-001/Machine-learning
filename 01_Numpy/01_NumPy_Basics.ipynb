# WORK to do this game and 
# define a mode funtion to get the most repeated number using unique function.

```python
import numpy as np
l=['Stone','Paper','Scissor']
user_point=0
computer=0
count=0
Rand_SPS=np.random.choice(l)

while count<3:
    user=input("Enter Stone or Paper or Scissor")
    count+=1
    if user=='Stone' and Rand_SPS=='Paper':
        computer+=1
    elif user=="Stone" and Rand_SPS=='Scissor':
        user_point+=1
    elif user=="Stone" and Rand_SPS=="Stone":
        print("Tie")
    elif user=="Paper" and Rand_SPS=='Scissor':
        user_point+=1
    elif user=="Paper" and Rand_SPS=="Stone":
        user_point+1
    elif user=="Paper" and Rand_SPS=='Paper':
        print("Tie")
    elif user=="Scissor" and Rand_SPS=="Stone":
        computer+=1
    elif user=="Scissor" and Rand_SPS=="Paper":
        user_point+=1
    elif user=="Scissor" and Rand_SPS=="Scissor":
        print("Tie")
if user_point>computer:
    print("YOU WON !!!!!")  
    print(f" User Point:{user_point}\n Computer Point:{computer}")
else:
    print("Computer WON!!!!!")
    print(f" User Point:{user_point}\n Computer Point:{computer}")
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
#print(x)
X_mean=x.mean()
Y_mean=y.mean()
len(x)

    
```

```python
Variance_X=0
Variance_Y=0
```

```python

for val in x:
    Variance_X+=(val-X_mean)**2/len(x)
```

```python

for val in y:
    Variance_Y+=(val-Y_mean)**2/len(y)
```

```python
print(Variance_X)
print(Variance_Y)
```

```python
cov_sum=0
for i in range(len(x)):
    cov_sum+=(x[i]-X_mean)*(y[i]-Y_mean)
cov_xy=cov_sum/len(x)
```
