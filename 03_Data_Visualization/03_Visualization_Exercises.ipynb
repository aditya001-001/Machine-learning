```python
playstore=playstore.dropna(subset=['Rating','Type','Price'])
```

```python
playstore['Installs']
# There are + and , in this column so we should remove it so pandas understands it as numeric columns
```

```python
playstore['Installs']=playstore['Installs'].str.replace(',','')
playstore['Installs']=playstore['Installs'].str.replace('+','')
playstore = playstore[playstore['Installs'] != 'Free']
```

```python
playstore['Installs'] != 'Free'
```

```python
playstore[playstore['Installs'] != 'Free']
```

```python
playstore['Installs'].unique
```

```python
playstore['Installs']=playstore['Installs'].astype(int)
```

```python
playstore['Price'] = playstore['Price'].str.replace('$', '')
playstore['Price'] = playstore['Price'].astype(float)
```

```python
playstore['Last Updated']=pd.to_datetime(playstore['Last Updated'])
playstore['year']=playstore['Last Updated'].dt.year
```

## 1.How has the average app rating changed over the years

```python
import matplotlib.pyplot as plt
grouped_year=playstore.groupby('year')
```

```python
rating=playstore.groupby('year')['Rating'].mean().reset_index()
plt.plot(rating['year'],rating['Rating'],marker='o')
plt.title('The average app rating changed over the years')
plt.xlabel('Years')
plt.ylabel('Rating')
plt.grid()
# It was high 4.2 in 2011 and fallen to 3.8 in 2012 then it improved and growing from 2017(4.1) to 2018 
```
