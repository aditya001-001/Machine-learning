* What is lasso(l1)?  
  * It is a regularization penalty which selects important features by pushing the weights towards to zero and exactly to the zero
  * If the weight is zero which means, that feature is not important 
* When to use l1?    
  * When we have more features.L1(lasso) 
* What is RIdge(L2)?    
  * It is a regularization penalty which keeps all the columns by strincking the weights 
  * It strinks the weights but never pushes to Zero 
* When to use Ridge(l2)?  
 * when all the feature columns are important 
 * When multi colinearity problem occurs 
* What is elastic net(l1+l2)?   
  * It is a regularization penalty which uses both l1 and l2  
  * it selects important features and also handles multi co-linearity problem 
* What is Gradient Descent?   
  * It is an optimization technique which is used to find the best weights 
* Where we use gradient Descent in ML?   
  * Logistic regression internally uses gradient descent to find the best weights  
  * Exstream gradient boosting algoritm (XGboost) uses gradient descent to find the best weights  
  * Regulirization techniques uses gradient Descent to find the best weights and reduces the variance.
