# Decision Tree Classifier and Regressor

Decision Trees are versatile machine learning algorithms that can perform both classification and regression tasks. This notebook covers:
*   How Decision Trees split nodes (Criterion: Gini Impurity vs. Entropy).
*   Training `DecisionTreeClassifier` and `DecisionTreeRegressor`.
*   Handling categorical splits on datasets.

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
```

```python
model1=DecisionTreeRegressor()
model2=DecisionTreeClassifier(criterion='gini',random_state=42)


```

```python
df=pd.read_csv("heart.csv")
```

```python
x=df.drop(columns='target')
y=df['target']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model2.fit(xtrain,ytrain)
```

```python
y_pred=model2.predict(xtest)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy=accuracy_score(ytest,y_pred)
print(accuracy)
```

```python
print(classification_report(ytest,y_pred))
```
