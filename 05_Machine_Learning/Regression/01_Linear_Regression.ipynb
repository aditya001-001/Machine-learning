```python
sam=pd.DataFrame({
    'study_hours':[1,3,5,7,2],
    'exam_score':[25,35,45,55,20]
})
```

```python
sam.info()
```

```python
X_mean=sam['study_hours'].mean()
Y_mean=sam['exam_score'].mean()
```

```python

for i in sam['study_hours']:
    sam['Z']=(i-X_mean)**2
```

```python
sam['Z']
```

```python
sam['Z'].sum()
```

```python
for i in sam['exam_score']:
    sam['V']=(i-Y_mean)**2
```

```python
sam['V']
```

```python
sam['V'].sum()
```

```python
print(sam['Z'],sam['V'],end=" ")
```

```python
sam['B']=sam['Z']*sam['V']
```

```python
sam['B'].sum()
```

```python
sam['Z'].sum()
```

```python
slope_numerator=sam['B'].sum()
slope_denominator=sam['Z'].sum()
```
